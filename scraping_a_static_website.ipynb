```python
#import BeautifulSoup
import bs4
import requests
import pandas as pd
```

```python
webpage = "https://en.wikipedia.org/wiki/U15_Canada"
```

```python
headers = {
    'User-Agent': "Mozilla/5/0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/120.0.0.0 Safari/537.36"
}
```

```python
web_page = bs4.BeautifulSoup(requests.get(webpage, headers=headers).text, 'lxml')
```

```python
web_page
```

```python
tables = web_page.find_all(name = 'table', attrs = {'class' : 'wikitable'})
```

```python
tables
```

```python
table = tables[-1]
table
```

```python
data = [[cell.get_text(strip = True) for cell in cells]
        for row in table.find_all('tr')
        for cells in [row.find_all(['td', 'th'])]
        ]
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```
